```python
import os
import pandas as pd

import warnings
warnings.filterwarnings("ignore")

# Path to folder containing the data
dataset_path = os.path.join(os.getcwd(), "datasets")

# Loading the data
df = pd.read_csv(os.path.join(dataset_path, "lego_sets.csv"))
parent_theme = pd.read_csv(os.path.join(dataset_path, "parent_themes.csv"))
```

```python
# Looking at lego sets data
df.head()
```

```python
# Looking at parent theme data
parent_theme.head()
```

```python
# Merging the two Datasets
merged_df = df.merge(parent_theme, left_on="parent_theme", right_on="name")

# Dropping the name_y column
merged_df.drop(columns="name_y", inplace=True)
merged_df.head()
```

### # Task 1
<p>What percentage of all licensed sets ever released were Star Wars themed? Save your answer as a variable <code>the_force</code> in the form of an integer (e.g. 25).</p>

```python
# Getting the records with is_licensed == True
licensed = merged_df[merged_df["is_licensed"]]

# Removing the all the records with empty set_num value
licensed.dropna(subset=["set_num"], inplace=True)
licensed.head()
```

```python
# Getting the records with parent themes as Star Wars 
star_wars = licensed[licensed["parent_theme"] == "Star Wars"]
star_wars.head()
```

```python
the_force = round(star_wars.shape[0]/licensed.shape[0]*100)
the_force
```

### # Task 2
<p>In which year was Star Wars not the most popular licensed theme (in terms of number of sets released that year)? Save your answer as a variable <code>new_era</code> in the form of an integer (e.g. 2012)</p>

```python
# Sorting the data by year
licensed_soted = licensed.sort_values("year")

# Grouping the data by year and parent_theme column. Then taking the sum of grouped columns and reseting the index of data
summed_df = licensed_soted.groupby(by=["year", "parent_theme"]).sum().reset_index()

# Sorting the records in descending order by is_licensed column and dropping the duplicate years
max_df = summed_df.sort_values("is_licensed", ascending=False).drop_duplicates(["year"])

# Againg sorting the values by year
max_df.sort_values("year", inplace=True)
max_df
```
